**Question 1: Download and combine all earthquake data from Jan 1980 until Jan 2020 into one dataframe. You can also get data all the way back to the 1970s if you prefer. Create a density map with a colorbar. Please also provide your code for creating your dataframe. (10 points)**

```python
import requests
import json
import geopandas as gpd
import pandas as pd
import cartopy.crs as crs
import matplotlib.pyplot as plt
```

```python
import pandas as pd
base_url = 'https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime='
df_list = []

for k in range (1980, 2021):
    for i in range (1,13):
        url2 = base_url + '%02d-%02d-01&endtime=%02d-%02d-01&minmagnitude=2.5' % (k,i,k,i+1)
        if i == 12:
            url2 = base_url + '%02d-%02d-01&endtime= %02d-01-01&minmagnitude=2.5' % (k,i,k+1)
        #print(url2)
        r2 = requests.get(url2)
        #print(r2.text)
        data = json.loads(r2.text)
        #print(data)
        crs = {'init': 'epsg:4326'}
        gdf = gpd.GeoDataFrame.from_features(data, crs=crs)
        df_list.append(gdf)
        print(url2, 'Finished!')
    #print(df_list)
```

```python
#Then, concatenate the data:
full_gdf1 = pd.concat(df_list, ignore_index=True)
```

```python
full_gdf1.to_file('earthquakes.shp')
```

```python
def colorbar_cartopy(mappable, cmap, **kwargs):
    from mpl_toolkits.axes_grid1 import make_axes_locatable
    ax = mappable.axes
    fig = ax.figure
    divider = make_axes_locatable(ax)   
    cax = divider.append_axes("right", size="1%", pad=0.05, axes_class=plt.Axes, autoscale_on=False)
    obj = fig.colorbar(cmap, cax=cax, **kwargs)
    return obj
```

**The top ten cities for earthquake magnitudes and their total magnitude from Jan 1980 until Jan 2020**

```python
pop_gdf = gpd.read_file(pop_places)
pop_gdf = pop_gdf[['NAME', 'geometry']]
```

```python
radius=0.5
```

```python
def total_mag(pop,quakes):
    pop['geometry'] = pop.geometry.buffer(radius)
    points = gpd.sjoin(quakes[['geometry','mag']], pop, how='inner', op='intersects')
    magsum= points.groupby('NAME').sum()
    high_magsum=magsum.sort_values(by='mag',ascending=False)
    return high_magsum.head(10)   
```

```python
cities_mag_sum=total_mag(pop_gdf,full_gdf)  
cities_mag_sum
```

**The top ten cities for earthquake frequency and their total number**

```python
pop_gdf = gpd.read_file(pop_places)
pop_gdf = pop_gdf[['NAME', 'geometry']]
```

```python
def frequency (pop,quakes):
    pop['geometry'] = pop.geometry.buffer(radius)
    f_cities = gpd.sjoin(pop[['geometry','NAME']], quakes, how='inner', op='intersects')
    counts_f = f_cities.groupby(['NAME']).count()
    counts_f_s=counts_f.sort_values(by='index_right',ascending=False)
    return counts_f_s.head(10)
```

```python
cities_frequency=frequency(pop_gdf,full_gdf)
cities_frequency
```

**Question 3: Using the gridded geodataframe we have developed, create maps showing (1) total magnitude of all earthquakes in each box, (2) total magnitude of all earthquakes above magnitude 5 in each box, (3) count of all earthquakes within each box, and (4) count of all earthquakes above magnitude 5 in each box. Which regions have the most high-magnitude earthquakes? (25 points).**

```python
import numpy as np
minX, maxX, minY, maxY, width = -180, 180, -60, 60, 0.25
Xspace = np.arange(minX, maxX + width, width)
Yspace = np.arange(minY, maxY + width, width)
```

```python
from shapely.geometry import Polygon
boxlist = []
for i, xval in enumerate(Xspace[:-1]):
    for j, yval in enumerate(Yspace[:-1]):
        box = [xval, Xspace[i+1], yval, Yspace[j+1]]
        box_coords = [[box[0], box[2]], [box[1], box[2]], [box[1], box[3]], [box[0], box[3]]]
        geom = Polygon([[p[0], p[1]] for p in box_coords])
        boxlist.append(geom)
boxes = gpd.GeoDataFrame(crs=crs, geometry=boxlist)
#boxes.head()
```
